```python
# Importamos las librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la XOR
X = np.array([[0,0,1,1], [0,1,0,1]]).T
y = np.array([0,1,1,0]).reshape(4,1)
```

#### a- Implementar SGD para resolver la XOR

```python
# Inicialización Random de los parámetros entre los valores 0.2 y 0.9 (para evitar valores muy bajos)
b1 = 0
b2 = 0
b3 = 0
np.random.seed(seed=1)
w11, w12, w21, w22, w31, w32 = np.random.uniform(0.2,0.9,6)
epochs = 5000
lr = 0.05
```

```python
def Sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

```python
# Implementamos las funciones de forward y backpropagation 
for epoch in range(epochs):
    l = 0
    for i in range(4):
        #Forward propagation
        z11 = w11*X[i][0] + w12*X[i][1] + b1
        z12 = w21*X[i][0] + w22*X[i][1] + b2
        a11 = Sigmoid(z11)
        a12 = Sigmoid(z12)
        z3 = w31*a11 + w32*a12 + b3
        y_hat = z3
        error = y[i] - y_hat
        l = l + np.power(error,2)
        #Back propagation
        dl_db3 = -2*error
        dl_dw31 = -2*error*a11
        dl_dw32 = -2*error*a12
        dl_db1 = -2*error*w31*a11*(1-a11)
        dl_dw11 = -2*error*w31*a11*(1-a11)*X[i][0]
        dl_dw12 = -2*error*w31*a11*(1-a11)*X[i][1]
        dl_db2 = -2*error*w32*a12*(1-a12)
        dl_dw21 = -2*error*w32*a12*(1-a12)*X[i][0]
        dl_dw22 = -2*error*w32*a12*(1-a12)*X[i][1]       
        
        
        # Actualizamos los parámetros
        w11 = w11 - lr*dl_dw11
        w12 = w12 - lr*dl_dw12
        w21 = w21 - lr*dl_dw21
        w22 = w22 - lr*dl_dw22
        w31 = w31 - lr*dl_dw31
        w32 = w32 - lr*dl_dw32
        w32 = w32 - lr*dl_dw32
        b1 = b1 - lr*dl_db1
        b2 = b2 - lr*dl_db2
        b3 = b3 - lr*dl_db3
    L = np.sqrt(l)
    print(L)
print('parametros: \n w11= {} \n w12= {} \n b1= {}'.format(w11,w12,b1))
print('parametros: \n w21= {} \n w22= {} \n b2= {}'.format(w21,w22,b2))
print('parametros: \n w31= {} \n w32= {} \n b3= {}'.format(w31,w32,b3))
```

#### d. Una vez entrenado el modelo, realizar predicciones para verificar el funcionamiento.

```python
# Predecimos con los valores de X la salida Y:

for i in range(4):
    z11 = w11*X[i][0] + w12*X[i][1] + b1
    z12 = w21*X[i][0] + w22*X[i][1] + b2
    a11 = Sigmoid(z11)
    a12 = Sigmoid(z12)
    z3 = w31*a11 + w32*a12 + b3
    y = z3
    print('X1: {} - X2: {} - Y: {} \n'.format(X[i][0],X[i][1],np.round(y,3)))
```
